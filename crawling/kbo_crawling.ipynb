# 네이버 스포츠 뉴스 크롤링

### Selenium 써보기
> 크롤링 끝날때마다 end_date가 'end_date.txt'로 저장됨<br>
크롤링 시작할때 'end_date.txt'를 읽어서 start_date에 넣어줌

```python
from selenium import webdriver
from datetime import datetime, timedelta
import time
import csv, pandas as pd
```

```python
chromedriver = 'driver/chromedriver'
driver = webdriver.Chrome(chromedriver)
```

```python
start_date = datetime(2019, 6, 26)
end_date = datetime(2020, 7, 30)

## 어디서 크롤링이 되다가 끊긴 상태면 'marked_date.txt'에서 marked_date 읽어오고, 없으면 처음 돌리는 것이므로 marked_date = start_date
import os.path

f_marked_date = 'marked_date.txt'

if os.path.isfile(f_marked_date):
    print('=== [marked_date.txt] exists ===')
    
    with open(f_marked_date, 'r') as f:
        marked_date = f.read()
        marked_date = datetime.strptime(marked_date, '%Y-%m-%d %H:%M:%S')
     
else:
    print('[marked_date.txt] does not exists...')
    
    marked_date = start_date
    
## marked_date 통해서 남은 크롤링 일수 계산
left_days = (end_date - marked_date).days
   
print(f'start_date = {start_date}')
print(f'end_date = {end_date}')
print(f'marked_date = {marked_date}')
print(f'left_days = {left_days}')
```

```python
## result 결과 파일 마련
f_results = 'results.csv'

if not os.path.isfile(f_results):
    
    with open(f_results, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)

        colname = ['date', 'num', 'url', 'title', 'content']

        writer.writerow(colname)
        
    # 처음 크롤링이니까 news_num = 1
    news_num = 1
        
    print(f'{f_results} 파일이 존재하지 않아 새로 만들었습니다.\n{news_num}번째 기사부터 크롤링합니다...')
    
    
else:
    # 이어서 크롤링이니까 csv 파일 읽어서 news_num 설정
    news_num = len(pd.read_csv('results.csv')) + 1
    
    print(f'===> {f_results} 파일이 존재하여 새로 만들지 않습니다.\n{news_num}번째 기사부터 크롤링합니다...') 
```

```python
crawling_days = 29 # 며칠 크롤링
for i in range(crawling_days):
    
    try:
        
        marked_date_forUrl = f'{marked_date:%Y%m%d}'
        
        print(f'** marked_date_forUrl = {marked_date_forUrl}')
        
        page_num = 1

        before_kbo_newslist = []
        kbo_newslist = []
        kbo_linkslist = []
        
        

        while(1) :

            kbo_url = f'https://sports.news.naver.com/kbaseball/news/index.nhn?isphoto=N&date={marked_date_forUrl}&page={page_num}'

            print(f'kbo_url = {kbo_url}')
            
            try: 
                driver.get(kbo_url)

            except:
                driver = webdriver.Chrome(chromedriver)
                driver.get(kbo_url)

            time.sleep(1)
            kbo_news_today = driver.find_element_by_class_name('news_list').find_elements_by_class_name('text')

            kbo_newslist = []
            for e in kbo_news_today:
                kbo_newslist.append(e.find_element_by_tag_name('a').get_attribute('href'))


            ## kbo_newslist를 반복적으로 돌면서 제목, 본문 추출
            for url in kbo_newslist:
                try: 
                    driver.get(url)

                except:
                    driver = webdriver.Chrome(chromedriver)
                    driver.get(url)
                    
                title = driver.find_element_by_class_name('title')
                content = driver.find_element_by_class_name('news_end')

                print(f'{news_num}번 기사 ===> title\n{title.text}\n\n')


                ## 크롤링한 데이터 저장
                with open(f_results, 'a+', newline='', encoding='utf-8') as f:
                    writer = csv.writer(f)
                    row = [marked_date_forUrl, str(news_num), url, title.text, content.text]
                    writer.writerow(row)
                    
                news_num += 1
                        
            

            if before_kbo_newslist != kbo_newslist : 
                kbo_linkslist += kbo_newslist
                before_kbo_newslist = kbo_newslist.copy()
                page_num += 1
                continue 

            break
            
        marked_date = marked_date + timedelta(days=1)
        
        
        
    except:
            
        print(f'ERROR: {str(marked_date)} 날짜 기사를 크롤링하는 도중 끊겼습니다.')
    

        ## marked_date부터 크롤링하면 되도록 marked_date 날짜 기사 크롤링했던거 다 삭제하기
        df = pd.read_csv(f_results, index_col=0)
        df = df.drop(f'{marked_date:%Y%m%d}')
        df.to_csv(f_results, sep=',', na_rep='NaN')
        
    
        driver.quit()
            
        break
        
## 어디까지 크롤링했는지 날짜를 저장
print(f'{str(marked_date)} 날짜 기사부터 크롤링하면 됩니다! [marked_date.txt] 파일에 어디까지 크롤링했는지 날짜를 저장합니다...')
with open(f_marked_date, 'w') as f:
    f.write(str(marked_date))
```
