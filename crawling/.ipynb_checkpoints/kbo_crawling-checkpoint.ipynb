# 네이버 스포츠 뉴스 크롤링

### Selenium 써보기
> 크롤링 끝날때마다 end_date가 'end_date.txt'로 저장됨<br>
크롤링 시작할때 'end_date.txt'를 읽어서 start_date에 넣어줌

```python
from selenium import webdriver
import datetime 
import time
```

```python
chromedriver = 'driver/chromedriver'
driver = webdriver.Chrome(chromedriver)
```

```python
# end_date를 현재 시간으로 맞춰줌
date = datetime.datetime.now() 
end_date = f'{date:%Y%m%d}'
```

```python
end_date = 20200728
page_num = 1

before_kbo_newslist = []
kbo_newslist = []
kbo_linkslist = []

while(1) :
    
    tmp = f'https://sports.news.naver.com/kbaseball/news/index.nhn?isphoto=N&date={end_date}&page={page_num}'
    
    try: 
        driver.get(tmp)
        
    except:
        driver = webdriver.Chrome(chromedriver)
        driver.get(tmp)
        
    time.sleep(1)
    kbo_news_today = driver.find_element_by_class_name('news_list').find_elements_by_class_name('text')
    
    kbo_newslist = []
    for e in kbo_news_today:
        kbo_newslist.append(e.find_element_by_tag_name('a').get_attribute('href'))
    
    if before_kbo_newslist != []:
        print(f'page_num = {page_num}\n\nbefore_kbo_newslist = {before_kbo_newslist[0]}\n\nkbo_newslist = {kbo_newslist[0]}\n\n')
    
    
    if before_kbo_newslist != kbo_newslist : 
        kbo_linkslist += kbo_newslist
        before_kbo_newslist = kbo_newslist.copy()
        page_num += 1
        continue 
    
    break
```
