### 기사 data 읽어오기

```python
import glob

total_news = 0
for filename in glob.glob('../kbo_data/*.csv'):
    df = pd.read_csv(filename, index_col=1, engine='python')
    len_df = len(df)
    total_news += len_df
    print(f'{filename}의 기사 개수 = {len_df}')
    
print(f'* 총 기사 개수 = {total_news}')
```
